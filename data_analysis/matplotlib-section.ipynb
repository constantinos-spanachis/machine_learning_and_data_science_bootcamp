## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare the data

x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setupl plot

fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data

ax.plot(x,y)

# 4. Customize Plot

ax.set(title= "Simple Plot",
            xlabel= "x-axis",
            ylabel = "y-axis")

# 5. Save and show

fig.savefig("./images/sample_plot.png")
```

# Bar plot

```python
# Make a plot from a dictionary

nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots();
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Prices",
      xlabel="Butter name",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()

ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

## Histogram

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots

```python
# 1. Method one

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10,10))

ax1.plot(x, x/2);

ax2.plot(np.linspace(0,1000,100), np.linspace(0,1000,100) ** 2);

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());

ax4.hist(x);
```

```python
# Option 2

fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0,0].scatter(x, np.cos(x));

ax[0,1].scatter(x, x**3);

ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());

ax[1,1].hist(np.random.uniform(0,1000, 500));
```

## Which one should you use? (Pyplot vs Matplotlib OO method?)

- When plotting something quickly, it is okay to use the pyplot method
- When plotting something more advanced, you should use the OO method

```python
over_50 = heart_disease.loc[heart_disease.age > 50]
```

```python
over_50
```

```python
# Pyplot method

over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax)
ax.set_xlim([45,over_50.age.max()+10])
```

```python
fix, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"])
# ax.set_xlim([45,over_50.age.max()+10])
ax.set(title="Heart Disease and Cholesterol Levels for people over 50",
      xlabel="Age",
      ylabel="Cholesterol")
# adding legend

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(y = over_50.chol.mean(),linestyle="--", c="black")

ax.axvline(x=over_50.age.mean(),linestyle="--", c="black")
```

```python
# subplot of chole , age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,8), sharex=True)

scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
)

ax0.set(title="Heart Disease and Cholesterol Levels for people over 50",
      ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y = over_50.chol.mean(), linestyle="--", c="black")

ax0.axvline(x=over_50.age.mean(), linestyle="--", c="black")

scatter = ax1.scatter(x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"])

ax1.set(title="Heart Disease and Thalach Levels for people over 50",
      xlabel="Age",
      ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y = over_50.thalach.mean(), linestyle="--", c="blue")

ax1.axvline(x=over_50.age.mean(), linestyle="--", c="blue")

fig.suptitle("Over 50 Heart Disease Analysis", fontsize=16, fontweight="bold" );
```

```python
plt.style.use('seaborn-whitegrid')

# subplot of chole , age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,8), sharex=True)

scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
    cmap="winter"
)

ax0.set(title="Heart Disease and Cholesterol Levels for people over 50",
      ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(y = over_50.chol.mean(), linestyle="--", c="red")

ax0.axvline(x=over_50.age.mean(), linestyle="--", c="red")

ax0.set_xlim([50,80])

ax0.set_ylim([100,600])

scatter = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"],
    cmap="winter"
)

ax1.set(title="Heart Disease and Thalach Levels for people over 50",
      xlabel="Age",
      ylabel="Max Heart Rate")

ax1.set_xlim([50,80])

ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(y = over_50.thalach.mean(), linestyle="--", c="blue")

ax1.axvline(x=over_50.age.mean(), linestyle="--", c="blue")

fig.suptitle("Over 50 Heart Disease Analysis", fontsize=16, fontweight="bold" );
```
